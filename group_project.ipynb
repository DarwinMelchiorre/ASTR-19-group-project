```python
import matplotlib.pyplot as plt
import csv

X = []
Y = []

with open('ASTR19_S22_group_project_data.txt', 'r') as datafile:
    plotting = csv.reader(datafile, delimiter=',')
    for ROWS in plotting:
        X.append(int(ROWS[0]))
        Y.append(int(ROWS[2]))

plt.plot(X, Y)
plt.title('Data from Text File')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
